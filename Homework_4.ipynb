# Question 2.1
# What is the impact of religiousness/spirituality on perceptions of non-traditional sexuality?

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns

# Select the relevant columns for analysis
data = alldata[[alldata.columns[5], alldata.columns[19]]]
# Perform the Chi-square test on contingency table
contingency_table = pd.crosstab(data[data.columns[0]], data[data.columns[1]])
# Significance level set to 5%
a = 0.05
(statistic, pvalue, dof, expected_frequency) = sp.stats.chi2_contingency(contingency_table, correction=True, lambda_=None)
if pvalue < a:
  print("Reject null hypothesis")
else:
  print("Fail to reject null hypothesis")
```

```python
# Create a heatmap of the contingency table
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')

# Set axis labels and plot title
plt.xlabel('Opinion on LGTBQ Question')
plt.ylabel('Religiousness/Spirituality')
plt.title('Heatmap of Opinions vs. Spirituality')

# Show the plot
plt.show()
```
